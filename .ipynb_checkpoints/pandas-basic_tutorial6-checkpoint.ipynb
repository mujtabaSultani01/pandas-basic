### Handling Missing Data using *'**replace()**'*' Method

```python
# First of all we import pandas and numpy, and read the CSV data:
import pandas as pd
import numpy as np
df = pd.read_csv("CSV Files/weather_Data1.csv")
df
```

```python
# So first of all, here via replace and numpy we can replace the missing values. -99999 is also missing values, much time such form of missing data will came in our datasets.
new_df = df.replace(-99999, np.NaN)
new_df
```

```python
# If we have multiple missing data, for example beside of -99999 if we have -88888, then we can replace both of them.
new_df = df.replace([-99999, -88888], np.NaN)
new_df 
```

```python
# So in upper case, we want to change the 0 values to NaN, so we should add it to the list with -99999 and -88888, as result if we have 0 values in teperature or windspeed columns then it will be also changed. To avoid it, we should assign Nan valuse to each column separately.
# See the the bellow approach:
new_df = df.replace({
    "temperature": -99999,
    "windspeed": [-88888, -99999],
    "event": '0'
}, np.NaN)
new_df
```

Yesssssss!!! it works fine.

```python
# We can also use mapping to replace specific values:
new_df = df.replace({
    -99999: np.NaN,
    -88888: np.NaN,
    '0': "Sunny"
})
new_df
```

Thats were all about replace method. for more necessary properties of replace method, you can search pandas documentation...
