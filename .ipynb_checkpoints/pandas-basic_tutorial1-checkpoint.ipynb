#### Data Frame (df) basics in pandas:
Data frame is a main object in pandas. it is used to represent data with rows and columns (Tabular or excell spread sheet like data). Tabular is a collection of rows and columns & we use it in pandas to represent data in tabular form.

```python
# read_csv() function read the csv files.
df = pd.read_csv('CSV Files/nyc_weather.csv')
df
```

```python
# head() function display the top n rows.
df.head(4)
```

```python
# tail() function display the bottom n rows.
df.tail(4)
```

```python
# We can create structured data using python dictionary:
all_electronic = {
    'name': ['Mobile', 'Laptop', 'Camera'],
    'price': ['140$', '580$', '180$'],
    'made in': ['US', 'UK', 'Chaina']
}
df1 = pd.DataFrame(all_electronic)
df1
```

```python
# To check the shape of dataframe, we use shape() function. It shows us the number of rows and columns.
df.shape
```

```python
df1.shape
```

```python
# We can assigns the rows and columns number to variables.
rows, columns = df.shape
```

```python
# rows stored the number of row.
rows
```

```python
# columns stored the number of columns.
columns
```

```python
# We can display a specific rows from DataFrame.
df[3:10]
```

```python
# Didn't show you the first 20 rows.
df[20:]
```

```python
# Show you the first 3 rows:
df[:3]
```

```python
# To display a specific columns:
df1[['price']]
```

```python
# To display a specific rows and columns:
df_sub = df[['EST', 'Humidity', 'DewPoint', 'Events', 'WindDirDegrees']]
df_sub.head(3)
```

```python
# To display a column type (Pandas.core.series.Series):
type(df1['name'])
```
