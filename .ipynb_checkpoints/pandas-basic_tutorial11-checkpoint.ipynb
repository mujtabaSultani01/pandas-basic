## Pandas Melt() Function
Similar to Pivot_table, melt function is used to transform or reshape the data.

```python
import pandas as pd1
df = pd.read_csv("CSV Files/weather4.csv")
df
```

```python
# So using pandas.melt function we can change the shape of the DataFrame:
df1 = pd1.melt(df, id_vars = ["day"])
df1.head(10)
```

```python
# We can choose a specific city in the same shape: 
df1[df1["variable"] == "chicago"]
```

```python
# Or 
df1[df1["variable"] == "chennai"]
```

```python
# We can also change the names of the column using vau_name and value_name properties:
df1 = pd1.melt(df, id_vars = ["day"], var_name = "city", value_name = "temperature")
df1.head(10)
```
